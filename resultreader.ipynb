```python
""" #CLEANUP
if len(sys.argv) < 3:
    print("please specify the input file (whole path)\n and the name of the output file.")
    sys.exit()
"""
#infile = sys.argv[1]
#infile = "../grid_search_1k/dump.txt"
#infile = "../grid_search_5k/dump.txt"
#infile = "../grid_search_10k/dump.txt"
#infile = "../grid_search_2_1k/dump.txt"
infile = "../grid_search_2_5k/dump.txt"


#outfile = "./log/results/"+sys.argv[2]
#outfile = "./log/results/grid_search_1k_results.csv"
#outfile = "./log/results/grid_search_5k_results.csv"
#outfile = "./log/results/grid_search_10k_results.csv"
#outfile = "./log/results/grid_search_2_1k_results.csv"
outfile = "./log/results/grid_search_2_5k_results.csv"



```

```python
run_to_config = {}
run_to_performance = {}

with open(infile, "r") as f:
    runcount = 0
    run = {} # temporary dict
    for line in f:
        if line == "\n":
            run_to_config[runcount]= run # add the block of specs to the internal representation of results
            runcount += 1
            run = {} # prepare for the next block of specs
        else: 
            line = line.rstrip()
            kv = line.split("\t")
            kv[0] = kv[0].rstrip() # get rid of spaces between key and value
            try:
                run[kv[0]] = float(kv[1]) # add key-value pair
            except ValueError:
                run[kv[0]] = kv[1] # for the flattening parameter
            if kv[0] == "sp_diff":
                print
                run_to_performance[runcount] = float(kv[1])
```

```python
ranked_runs = sorted(run_to_performance, key=run_to_performance.get, reverse=True)
with open(outfile, "w") as f:
    if   outfile[-3:] == "csv": sep = ","
    elif outfile[-3:] == "tsv": sep = "\t"
    elif outfile[-3:] == "txt": sep = "   \t"
    else:                       sep = " "
        
    f.write("score"+sep+"KC_factor"+sep+"projections"+sep+"hash_dims"+sep+"flattening"+"\n")
    for run in ranked_runs:
        line = str(run_to_performance[run])+sep+\
               str(run_to_config[run]["KC_factor"])+sep+\
               str(run_to_config[run]["projections"])+sep+\
               str(run_to_config[run]["hash_dims"])+sep+\
               str(run_to_config[run]["flattening"])+"\n"
        f.write(line)
        print(line.rstrip())
    print("number of runs: ",len(ranked_runs))
    
```
